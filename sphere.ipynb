## Import Libraries

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Declare Functions

```python
# Get points
def get_points(r,pieces):
  # Declare variables
  angle = 360/pieces
  angles = [angle*i for i in range(pieces + 1)]
  x1 = 0
  y1 = 0
  x2 = 0
  z2 = 0
  x_series = []
  y_series = []
  z_series = []
  # Get points
  for a in angles:  
    x1 = r*math.cos(a)
    y1 = r*math.sin(a)
    for a in angles:
      x2 = x1*math.cos(a)
      z2 = x1*math.sin(a)
      x_series.append(x2)
      y_series.append(y1)
      z_series.append(z2)
      # Validate distance
      assert round(x2**2+y1**2+z2**2,2) == round(r**2,2)

  return x_series,y_series,z_series

# Visualize
def show_sphere(x_series,y_series,z_series):

  fig = plt.figure(figsize=(12, 12))
  ax = fig.add_subplot(projection='3d')
  ax.set_xlabel('X-axis')
  ax.set_ylabel('Y-axis')
  ax.set_zlabel('Z-axis')
  ax.scatter(x_series, z_series, y_series )

  return plt
```

## Implement

```python
r = 10
pieces = [4,16,32,64,128]
for p in pieces:
  print("Pieces: ", p)
  x_series,y_series,z_series = get_points(r,p)
  show_sphere(x_series,y_series,z_series).show()
```
